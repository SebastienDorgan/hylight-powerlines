```python
import glob

from PIL import ExifTags, Image

images = []
root_dir = "../assets/Test - Powerlines pictures"
for image in glob.glob("*.JPG", root_dir=root_dir):
    images.append(image[-12:])
    img = Image.open(f"{root_dir}/{image}")
    exif = img.getexif()
    print(image)
    print(img.size)
    for tag_id, value in exif.items():
        tag = ExifTags.TAGS.get(tag_id, tag_id)
        print(f"{tag}: {value}")
    print("----")
assert len(images) == len(set(images))
```
